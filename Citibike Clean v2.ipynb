### Jersey City Stations & API Call

```python
#start stations
dfjs=df[['start_station_name', 'start_lat', 'start_lng']].drop_duplicates(subset= ['start_station_name'])
dfjs= dfjs.rename(columns={"start_station_name": "name", "start_lat": "lat", "start_lng": "lng"})
dfjs
```

```python
#end stations
dfje=df[['end_station_name', 'end_lat', 'end_lng']].drop_duplicates(subset= ['end_station_name'])
dfje= dfje.rename(columns={"end_station_name": "name", "end_lat": "lat", "end_lng": "lng"})
dfje
```

```python
#combine the two to be sure we have all stations
dfjall=pd.concat([dfjs,dfje]).drop_duplicates().reset_index(drop=True).dropna()
dfjall
```

```python
#add columns that will hold api data in df
dfjall['borough']=''
dfjall['zipcode']=''
```

```python
borough=[]
zipcode=[]
allcall=[]

for index, row in dfjall.iterrows():
   
         #lat and lon inside the loop so it only takes for that row and add it to search params  
    lat=row["lat"]
    lon=row["lng"]
    base_url = f"https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={lon}&format=json"
    # make request and collect the json into list
    response = requests.get(base_url).json()
    allcall.append(response)
    try:
        dfjall.loc[index, "borough"] =response['address']["city"]
    except: 
        dfjall.loc[index, "borough"] ='Jersey City'
    try:
        dfjall.loc[index, "zipcode"]=response['address']['postcode']
    except:
    
        time.sleep(.5)
```
